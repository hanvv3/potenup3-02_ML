````python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error, r2_score


df = pd.read_csv('./data/rider_data.csv')
df
````

````python

# 1. í•™ìŠµìš©(Train)ê³¼ í…ŒìŠ¤íŠ¸ìš©(Test) ë°ì´í„° ë¶„ë¦¬ (8:2)
X = df[['Distance_km', 'Riders_Available', 'Weather_Index', 'Rush_Hour']]
y = df['Time_min']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=3333)

# 2. ëª¨ë¸ ìƒì„± ë° í•™ìŠµ
model = LinearRegression()
model.fit(X_train, y_train)

# 3. ì˜ˆì¸¡ ì‹¤í–‰ (í…ŒìŠ¤íŠ¸ ë°ì´í„°ë¡œ ì‹œí—˜)
y_pred = model.predict(X_test)

# 4. ì„±ëŠ¥ í‰ê°€
r2 = r2_score(y_test, y_pred)
mae = mean_absolute_error(y_test, y_pred)

print(f"ëª¨ë¸ ì„¤ëª…ë ¥(R2 Score): {r2:.4f} (1.0ì— ê°€ê¹Œìš¸ìˆ˜ë¡ ì™„ë²½)")
print(f"í‰ê·  ì˜¤ì°¨(MAE): {mae:.2f}ë¶„ (ì˜ˆì¸¡ì´ í‰ê· ì ìœ¼ë¡œ ì•½ {mae:.2f}ë¶„ ë¹—ë‚˜ê°)")
````

````python
# íšŒê·€ ê³„ìˆ˜ë¡œ ì¤‘ìš”í•œ í”¼ì³ë“¤ì„ ê³¨ë¼ë‚´ì–´ ê°œì„ í•˜ê±°ë‚˜ ì •í™•ë„ë¥¼ ì˜¬ë¦´ ìˆ˜ ìˆë‹¤.
## ì¶œí‡´ê·¼ ì‹œê°„ì˜ ë„ë¡œìƒí™©ì˜ ë°ì´í„°/í”¼ì³ê°€ ë” ë§ì•„ì§„ë‹¤ë©´ ì •í™•ë„ë¥¼ ë” ì˜¬ë¦´ ìˆ˜ ìˆë‹¤.
model.coef_

coef_df = pd.DataFrame({
    "feature": X.columns,
    "Coef": model.coef_
})

print("y-intercept:", model.intercept_)

print("\n------ íšŒê·€ ê³„ìˆ˜ ------")
coef_df
````

````python
from sklearn.linear_model import Ridge, Lasso
from sklearn.model_selection import GridSearchCV

r_model = Ridge()
l_model = Lasso()


# 1. í…ŒìŠ¤íŠ¸í•  alpha ê°’ë“¤ì„ ë¦¬ìŠ¤íŠ¸ë¡œ ë§Œë“­ë‹ˆë‹¤. (ë³´í†µ ë¡œê·¸ ìŠ¤ì¼€ì¼ ì‚¬ìš©)
param_grid = {'alpha': [0.001, 0.01, 0.1, 1, 10, 100]}

# 2. ê·¸ë¦¬ë“œì„œì¹˜ ì„¤ì • (5-ê²¹ êµì°¨ ê²€ì¦ ì‚¬ìš©)
r_grid_search = GridSearchCV(r_model, param_grid, cv=5)

# 3. í•™ìŠµ (ì´ ê³¼ì •ì—ì„œ 6ê°œì˜ alphaë¥¼ ê°ê° í…ŒìŠ¤íŠ¸í•˜ë©° ìµœì ì˜ ê°’ì„ ì°¾ìŠµë‹ˆë‹¤)
r_grid_search.fit(X_train, y_train)

# 4. ê²°ê³¼ í™•ì¸
print(f"L2 ìµœì ì˜ alpha: {r_grid_search.best_params_}")
print(f"L2 ìµœê³  ì ìˆ˜: {r_grid_search.best_score_:.3f}")

# 5. ì˜ˆì¸¡ ì‹¤í–‰ (í…ŒìŠ¤íŠ¸ ë°ì´í„°ë¡œ ì‹œí—˜)
y_pred = r_grid_search.best_estimator_.predict(X_test)

# 6. ì„±ëŠ¥ í‰ê°€
r2 = r2_score(y_test, y_pred)
mae = mean_absolute_error(y_test, y_pred)

print(f"L2 ëª¨ë¸ ì„¤ëª…ë ¥(R2 Score): {r2:.4f} (1.0ì— ê°€ê¹Œìš¸ìˆ˜ë¡ ì™„ë²½)")
print(f"L2 í‰ê·  ì˜¤ì°¨(MAE): {mae:.2f}ë¶„ (ì˜ˆì¸¡ì´ í‰ê· ì ìœ¼ë¡œ ì•½ {mae:.2f}ë¶„ ë¹—ë‚˜ê°)")



# 2. ê·¸ë¦¬ë“œì„œì¹˜ ì„¤ì • (5-ê²¹ êµì°¨ ê²€ì¦ ì‚¬ìš©)
l_grid_search = GridSearchCV(l_model, param_grid, cv=5)

# 3. í•™ìŠµ (ì´ ê³¼ì •ì—ì„œ 6ê°œì˜ alphaë¥¼ ê°ê° í…ŒìŠ¤íŠ¸í•˜ë©° ìµœì ì˜ ê°’ì„ ì°¾ìŠµë‹ˆë‹¤)
l_grid_search.fit(X_train, y_train)

# 4. ê²°ê³¼ í™•ì¸
print(f"L1 ìµœì ì˜ alpha: {l_grid_search.best_params_}")
print(f"L1 ìµœê³  ì ìˆ˜: {l_grid_search.best_score_:.3f}")

# 5. ì˜ˆì¸¡ ì‹¤í–‰ (í…ŒìŠ¤íŠ¸ ë°ì´í„°ë¡œ ì‹œí—˜)
y_pred = l_grid_search.best_estimator_.predict(X_test)

# 6. ì„±ëŠ¥ í‰ê°€
r2 = r2_score(y_test, y_pred)
mae = mean_absolute_error(y_test, y_pred)

print(f"L1 ëª¨ë¸ ì„¤ëª…ë ¥(R2 Score): {r2:.4f} (1.0ì— ê°€ê¹Œìš¸ìˆ˜ë¡ ì™„ë²½)")
print(f"L1 í‰ê·  ì˜¤ì°¨(MAE): {mae:.2f}ë¶„ (ì˜ˆì¸¡ì´ í‰ê· ì ìœ¼ë¡œ ì•½ {mae:.2f}ë¶„ ë¹—ë‚˜ê°)")
````
