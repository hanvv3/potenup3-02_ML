### ğŸ·ï¸ LabelEncoder: ë²”ì£¼í˜• ë°ì´í„° ìˆ«ìë¡œ ë°”ê¾¸ê¸°

`LabelEncoder`ëŠ” ë¬¸ìë¡œ ëœ ì¹´í…Œê³ ë¦¬ ë°ì´í„°ë¥¼ ë‹¨ìˆœí•œ **ìˆ«ì ë²ˆí˜¸**ë¡œ ë°”ê¾¸ëŠ” ê°€ì¥ ê¸°ë³¸ì ì¸ ë„êµ¬ì…ë‹ˆë‹¤.

### í•µì‹¬ ìš”ì•½
* **ì—­í• **: ë²”ì£¼í˜•(ë¬¸ì) ë°ì´í„°ë¥¼ ì •ìˆ˜(0, 1, 2, ...)ë¡œ ë³€í™˜í•©ë‹ˆë‹¤.
* **ìˆœì„œ**: ê°€ë‚˜ë‹¤ ë˜ëŠ” ì•ŒíŒŒë²³ ìˆœì„œëŒ€ë¡œ 0ë²ˆë¶€í„° ë²ˆí˜¸ë¥¼ ë§¤ê¹ë‹ˆë‹¤.
* **ìš©ë„**: ì£¼ë¡œ ëª¨ë¸ì´ ë§ì¶°ì•¼ í•˜ëŠ” **ì •ë‹µ(Target) ë³€ìˆ˜**ë¥¼ ë³€í™˜í•  ë•Œ ì‚¬ìš©í•©ë‹ˆë‹¤.

```python
from sklearn.preprocessing import LabelEncoder

# 1. ë°ì´í„° ì¤€ë¹„
items = ['TV', 'ëƒ‰ì¥ê³ ', 'ì „ìë ˆì¸ì§€', 'ì»´í“¨í„°', 'ì„ í’ê¸°', 'ëƒ‰ì¥ê³ ', 'TV']

# 2. LabelEncoder ê°ì²´ ìƒì„± ë° í•™ìŠµ/ë³€í™˜
encoder = LabelEncoder()
labels = encoder.fit_transform(items)

# 3. ê²°ê³¼ í™•ì¸
print(f"ì¸ì½”ë”© ë³€í™˜ê°’: {labels}")
print(f"ì¸ì½”ë”© í´ë˜ìŠ¤(ìˆœì„œ): {encoder.classes_}")

# 4. ì—­ë³€í™˜ (ìˆ«ìë¥¼ ë‹¤ì‹œ ë¬¸ìë¡œ)
decoded = encoder.inverse_transform([4, 0, 2])
print(f"ìˆ«ì [4, 0, 2]ì˜ ì›ë³¸ê°’: {decoded}")
decoded
```

### âš–ï¸ í”¼ì²˜ ìŠ¤ì¼€ì¼ë§: í‘œì¤€í™” (Standardization)

**í‘œì¤€í™”**ëŠ” ë°ì´í„°ì˜ í‰ê· ì„ 0, í‘œì¤€í¸ì°¨ë¥¼ 1ë¡œ ë³€í™˜í•˜ì—¬ ë°ì´í„°ë¥¼ **í‘œì¤€ ì •ê·œ ë¶„í¬**ì˜ í˜•íƒœë¡œ ë§Œë“œëŠ” ê³¼ì •ì…ë‹ˆë‹¤.

### 1. í•µì‹¬ ìš”ì•½
* **ì—­í• **: ë°ì´í„°ê°€ í‰ê· ìœ¼ë¡œë¶€í„° ì–¼ë§ˆë‚˜ ë–¨ì–´ì ¸ ìˆëŠ”ì§€(í‘œì¤€í¸ì°¨ì˜ ëª‡ ë°°ë§Œí¼)ë¥¼ ë‚˜íƒ€ë‚´ëŠ” ìˆ˜ì¹˜ë¡œ ë³€í™˜í•©ë‹ˆë‹¤.
* **ê³µì‹**: $$z = \frac{x - \mu}{\sigma}$$ (ë‹¨, $\mu$ëŠ” í‰ê· , $\sigma$ëŠ” í‘œì¤€í¸ì°¨)
* **íŠ¹ì§•**: ì •ê·œí™”(Min-Max)ì™€ ë‹¬ë¦¬ ë°ì´í„°ì˜ ë²”ìœ„ê°€ ì œí•œë˜ì§€ ì•Šìœ¼ë©°, ë§ˆì´ë„ˆìŠ¤(-) ê°’ì´ ë°œìƒí•  ìˆ˜ ìˆìŠµë‹ˆë‹¤.



### ğŸ’¡ í”¼ì²˜ ìŠ¤ì¼€ì¼ë§ì˜ ì‘ë™ ë°©ì‹: "ì—´(Column) ë‹¨ìœ„ ê³„ì‚°"

ë°ì´í„° í”„ë ˆì„ì— ì—¬ëŸ¬ ê°œì˜ ì—´ì´ ìˆì„ ë•Œ, ìŠ¤ì¼€ì¼ëŸ¬ëŠ” ì „ì²´ ë°ì´í„°ë¥¼ í•˜ë‚˜ë¡œ ë­‰ì³ì„œ ê³„ì‚°í•˜ì§€ ì•Šê³  **ê° ì—´ì„ ë…ë¦½ì ì¸ ë‹¨ìœ„**ë¡œ ì·¨ê¸‰í•©ë‹ˆë‹¤.

* **ë…ë¦½ì  ê³„ì‚°**: `Age` ì—´ì€ `Age`ë¼ë¦¬, `Salary` ì—´ì€ `Salary`ë¼ë¦¬ ê°ê°ì˜ í‰ê· ê³¼ í‘œì¤€í¸ì°¨ë¥¼ êµ¬í•©ë‹ˆë‹¤.
* **ì´ìœ **: ê° ì—´ë§ˆë‹¤ ì¸¡ì • ë‹¨ìœ„(Unit)ê°€ ë‹¤ë¥´ê¸° ë•Œë¬¸ì…ë‹ˆë‹¤. (ì„¸ì›” vs í™”í ë‹¨ìœ„)
* **ê²°ê³¼**: ë³€í™˜ì´ ëë‚˜ë©´ ëª¨ë“  ì—´ì€ ê°ìì˜ ë¶„í¬ ì•ˆì—ì„œ í‰ê·  0, í‘œì¤€í¸ì°¨ 1(í‘œì¤€í™” ì‹œ)ì´ë¼ëŠ” **ê³µí†µëœ ê¸°ì¤€ì **ì„ ê°–ê²Œ ë˜ì–´ ëª¨ë¸ì´ ê³µì •í•˜ê²Œ í•™ìŠµí•  ìˆ˜ ìˆìŠµë‹ˆë‹¤.

```python
from sklearn.preprocessing import StandardScaler
import numpy as np
import pandas as pd

# 1. ë°ì´í„° ì¤€ë¹„ (ì˜ˆ: ë‚˜ì´ì™€ ì—°ë´‰)
data = {
    'Age': [25, 35, 45, 50],
    'Salary': [2500, 4800, 7100, 5200]
}
df = pd.DataFrame(data)

# 2. StandardScaler ê°ì²´ ìƒì„±
scaler = StandardScaler()

# 3. ë°ì´í„° í•™ìŠµ ë° ë³€í™˜ (fit + transform)
# fit: í‰ê· ê³¼ í‘œì¤€í¸ì°¨ë¥¼ ê³„ì‚° / transform: ì‹¤ì œ ê°’ ë³€í™˜
scaled_data = scaler.fit_transform(df)

# 4. ê²°ê³¼ í™•ì¸ (ì†Œìˆ˜ì  ë‘˜ì§¸ìë¦¬ê¹Œì§€)
scaled_df = pd.DataFrame(scaled_data, columns=df.columns)

print("--- ì›ë³¸ ë°ì´í„° ---")
print(df)
print("\n--- í‘œì¤€í™” í›„ ë°ì´í„° (í‰ê·  0, í‘œì¤€í¸ì°¨ 1) ---")
print(scaled_df.round(2))

# 5. í†µê³„ì¹˜ í™•ì¸
print("\n--- ë³€í™˜ í›„ í‰ê· ê°’ (ê±°ì˜ 0) ---")
print(scaled_df.mean().round(2))
print("\n--- ë³€í™˜ í›„ í‘œì¤€í¸ì°¨ (ê±°ì˜ 1) ---")
print(scaled_df.std().round(2))
```

# ğŸ“ í”¼ì²˜ ìŠ¤ì¼€ì¼ë§: ì •ê·œí™” (Min-Max Scaling)

**ì •ê·œí™”**ëŠ” ì„œë¡œ ë‹¤ë¥¸ ë²”ìœ„ì˜ ìˆ«ì ë°ì´í„°ë¥¼ **0ê³¼ 1 ì‚¬ì´ì˜ ì¼ì •í•œ ë²”ìœ„**ë¡œ ì••ì¶•í•˜ëŠ” ê³¼ì •ì…ë‹ˆë‹¤.

### 1. í•µì‹¬ ìš”ì•½
* **ì—­í• **: ë°ì´í„°ì˜ ìµœì†Ÿê°’ì„ 0, ìµœëŒ“ê°’ì„ 1ë¡œ ë³€í™˜í•˜ì—¬ ëª¨ë“  ë°ì´í„°ë¥¼ ì´ ë²”ìœ„ ì•ˆì— ë§ì¶¥ë‹ˆë‹¤.
* **ê³µì‹**: $$x_{scaled} = \frac{x - x_{min}}{x_{max} - x_{min}}$$
* **íŠ¹ì§•**: ë°ì´í„°ì˜ ìƒëŒ€ì ì¸ ìœ„ì¹˜ëŠ” ìœ ì§€í•˜ë©´ì„œ í¬ê¸°ë§Œ ì¤„ì…ë‹ˆë‹¤. ëª¨ë“  ì—´(Feature)ì´ ë™ì¼í•œ ìµœì†Œ/ìµœëŒ€ ë²”ìœ„ë¥¼ ê°–ê²Œ ë©ë‹ˆë‹¤.

[Image of Min-Max scaling formula and data points mapped from original scale to 0-1 range]

---

```python
from sklearn.preprocessing import MinMaxScaler
import pandas as pd
import numpy as np

# 1. ë°ì´í„° ì¤€ë¹„ (ê° ì—´ì˜ ë²”ìœ„ê°€ ì•„ì£¼ ë‹¤ë¥¸ ê²½ìš°)
# ì ìˆ˜: 0~100ì  / ì—°ë´‰: ìˆ˜ì²œë§Œ ì› ë‹¨ìœ„
data = {
    'Score': [95, 80, 70, 100],
    'Salary': [4500, 3000, 2800, 5000]
}
df = pd.DataFrame(data)

# 2. MinMaxScaler ê°ì²´ ìƒì„±
scaler = MinMaxScaler()

# 3. ë°ì´í„° í•™ìŠµ ë° ë³€í™˜ (ê° ì—´ë³„ë¡œ ìµœì†Ÿê°’ 0, ìµœëŒ“ê°’ 1 ì„¤ì •)
scaled_data = scaler.fit_transform(df)

# 4. ê²°ê³¼ í™•ì¸
scaled_df = pd.DataFrame(scaled_data, columns=df.columns)

print("--- ì›ë³¸ ë°ì´í„° ---")
print(df)
print("\n--- ì •ê·œí™” í›„ ë°ì´í„° (0~1 ë²”ìœ„) ---")
print(scaled_df)
```

```python
from sklearn.preprocessing import LabelEncoder
from sklearn.preprocessing import StandardScaler
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import GridSearchCV
import numpy as np
import pandas as pd
```

```python
# ë°ì´í„° ë¶ˆëŸ¬ì˜¤ê¸°
train_df = pd.read_excel("./data/train_dataset.xlsx")
test_df = pd.read_excel("./data/test_dataset.xlsx")


train_df.head()
```

```python
prep_train_df = train_df.copy()

cols_to_scale = ["Age", "Experience(Years)", "Salary(ë§Œì›)"]

scaler = StandardScaler()


prep_train_df[cols_to_scale] = scaler.fit_transform(train_df[cols_to_scale])

prep_train_df.head()
```

```python
X_train = prep_train_df.drop("Purchased", axis=1)
y_train = prep_train_df["Purchased"]

print(X_train.shape)
print(y_train.shape)
```

```python
prep_test_df = test_df.copy()

prep_test_df[cols_to_scale] = scaler.transform(test_df[cols_to_scale])

prep_test_df.head()
```

```python
X_test = prep_test_df.drop("Purchased", axis=1)
y_test = prep_test_df["Purchased"]

print(X_test.shape)
print(y_test.shape)
```

```python
model = DecisionTreeClassifier(random_state=42)

params = {"max_depth": [1,2,3,4,5,6,7], "min_samples_split": [2,3,4,5,6,7]}

grid_model = GridSearchCV(model, param_grid=params, cv=5)

grid_model.fit(X_train, y_train)
```

```python
score_df = pd.DataFrame(grid_model.cv_results_)
score_df.head()
```

```python
print(f"best params = {grid_model.best_params_}")
print(f"best estimator = {grid_model.best_estimator_}")
```

```python
pred = grid_model.best_estimator_.predict(X_test)

res = np.equal(y_test, pred)
res.sum()/res.size
```

```python
from sklearn.metrics import accuracy_score

print(accuracy_score(y_test, pred))

print(grid_model.best_estimator_.score(X_test, y_test))
```
