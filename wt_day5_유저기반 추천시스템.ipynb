# 1. 유저기반 추천시스템

```python
movies_df = pd.read_csv(
    "./data/practice/movies.csv"
)

movies_df.head()
```

```python
ratings_df = pd.read_csv(
    "./data/practice/ratings.csv"
)

ratings_df.head()
```

```python
movies_df.info()
```

```python
ratings_df.info()
```

```python
df = pd.merge(
    left=ratings_df,
    right=movies_df,
    how="inner",
    on="movieId"
)

df.head()
```

```python
df.info()
```

```python
user_ratings_df = df.pivot_table(
    index="userId",
    columns="title",
    values="rating"
)

user_ratings_df.head()
```

```python
user_ratings_df = user_ratings_df.fillna(0)
user_ratings_df.head()
```

```python
from sklearn.metrics.pairwise import cosine_similarity

user_base_result = cosine_similarity(user_ratings_df, user_ratings_df)
user_base_df = pd.DataFrame(
    data=user_base_result, 
    index=user_ratings_df.index,
    columns=user_ratings_df.index
)

user_base_df
```

```python
user_base_df.iloc[0].sort_values(ascending=False)   # 만약 유사도가 낮다면 다른 추천시스템을 도입해야 함

## 1번 유저기준 266번 유저의 유사도가 가장 높았지만 35.7%는 유사하다고 할 수 없다..
```

```python
## 0번 유저가 본 영화 리스트업하고 다른 가장 유사도가 높았던 유저(266)의 영화 리스트업한다. 
## 같은 영화를 제외하고 추천해주기

my_list = []

for idx, rating in enumerate(user_ratings_df.iloc[0]):
    
    if rating > 0:
        my_list.append(user_ratings_df.iloc[0].index[idx])


print(f"유저(0번)가 본 영화의 개수 : {len(my_list)}")
```

```python
u_data = user_ratings_df.iloc[266]

u_data[u_data>3].index
```

```python
# set()을 사용하여 차집합을 구한다: 다른유저가 본것들 중에 내가 본것을 뺀다.
# 응용할 수 있다

set(u_data[u_data>3].index) - set(my_list)
```

```python
# 응용할 수 있다: 
## 예시1) 유사도 상위 3명의 다른 유저들의 교집합을 구해서 내가 본것을 뺀다. <-- 해보기.
## 예시2) 결과로 나온 영화의 ratings의 평균(유사한 유저들의 평균 점수)을 구해서 추천
```

## 1) 유저기반 추천시스템 응용1
* 유사도 상위 3명의 다른 유저들의 교집합을 구해서 내가 본것을 뺀다.

```python
# 예시1)
top3_users = user_base_df.iloc[0].sort_values(ascending=False)[1:4].index.values
top3_users
```

```python
top3_data = user_ratings_df.iloc[top3_users-1]      # 인덱스 값이라 1 빼줌
top3_data
```

```python
# 내가 짠 코드
set_list = []

for u in top3_users:
    m = top3_data.loc[u,:]
    m = set(m[m>3].index)
    set_list.append(m)

common = set.intersection(*set_list) if set_list else set()         # 3명의 유저가 4점이상 평가한 영화들의 교집합
common
```

```python
# GPT 버전의 깔끔한 코드: pivot에서 바로 구하기
top3_rows = top3_data.loc[top3_users]          # shape: (3, n_movies)
common = set(top3_rows.columns[(top3_rows > 3).all(axis=0)])
common
```

```python
# top3 유저의 교집합에서 내가 본 것 빼기.
common - set(my_list)
```
